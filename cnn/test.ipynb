```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import pathlib
import keras as krs
from krs import layers, models
from tensorflow.python.keras import ImageDataGenerator
from tensorflow.python.keras import image
from sklearn.model_selection import train_test_split
```

```python
# ========== 1. Atur Path Dataset ==========
DATASET_PATH = "dataset"  # pastikan di folder yang sama dengan model_cnn.py
BATCH_SIZE = 16
IMG_SIZE = (128, 128)
EPOCHS = 10

# ========== 2. Preprocessing: Data Generator ==========
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% buat validasi
)

train_generator = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)

# ========== 3. Bangun Model CNN ==========
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(*IMG_SIZE, 3)),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')  # binary classification
])

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
# ========== 4. Train Model ==========
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator
)
```

```python
# ========== 5. Simpan Model ==========
for layer in model.layers:
    layer.trainable = False

model.save("model_daun_cnn.h5")
```
